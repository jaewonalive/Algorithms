```python
def solution(m, musicinfos):
    def make_note_list(note_string):
        j=0
        note_string_ls = []
        for k in note_string:
            j+=1
            
            if j == len(note_string):
                note_string_ls.append(k)
                break
            
            if (k in ['C', 'D', 'F', 'G', 'A']) & (note_string[j] == '#'):
                note_string_ls.append(k+'#')
                
            elif k == '#':
                continue
                
            else : 
                note_string_ls.append(k)
                    
        return note_string_ls    
    
    
    relevant_info = []
    for i in musicinfos:
        one_info = i.split(',')
        
        start_time_ls = one_info[0].split(':')
        end_time_ls = one_info[1].split(':')
        
        start_time_int = int(start_time_ls[0])*60 + int(start_time_ls[1])
        end_time_int = int(end_time_ls[0])*60 + int(end_time_ls[1])
        
        played_length = end_time_int - start_time_int
        del one_info[0:2]
        one_info.insert(0, played_length)
        
        tmp_num_sharp = one_info[2].count('#')
        tmp_note_info = one_info[2].replace('#','')
        
        
        if played_length <= len(tmp_note_info):
            played_note = one_info[2][:played_length+tmp_num_sharp]
            
        else :
            mok = played_length//len(tmp_note_info)
            tmp_note = one_info[2] * mok
            played_note = tmp_note[:played_length+tmp_num_sharp]
        
        one_info.append(played_note)
        
            
        played_note_ls = make_note_list(played_note)
        m_note_ls = make_note_list(m)
            
            
        w = 0
        while True :
            if w + len(m_note_ls) >= len(played_note_ls):
                lapsed_num = w + len(m_note_ls) - len(played_note_ls) 
                checked_note = ''.join(played_note_ls[w:]+\
                                       played_note_ls[:lapsed_num])
                if m == checked_note:
                    relevant_info.append(one_info)
                    break
                    
                if w >= len(played_note_ls)-1:
                    break
                    
                               
            checked_note = ''.join(played_note_ls[w:w+len(m_note_ls)])
            
            if m == checked_note:
                relevant_info.append(one_info)
                break
            
            w+=1
            
            
                
    if len(relevant_info) >= 1:
        relevant_info = sorted(relevant_info,
                        key=lambda one_info : one_info[0],
                        reverse = True)
    
        answer = relevant_info[0][1]
        
    else :
        answer ='(None)'
            

    return answer
```
